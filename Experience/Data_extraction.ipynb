```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
with open("Industrial_Giants.txt","r") as file:
    Giants_company_raw=file.read()
splitted_Giants_company=Giants_company_raw.split("\n")
df_giants=pd.DataFrame({"Companyname":splitted_Giants_company})
df_giants.drop_duplicates(inplace=True)
level=[]
for i in range(len(df_giants)):
    level.append("Giants")
df_giants["level"]=level
```

```python
with open("Major tech companies.txt","r") as file:
    major_data_raw=file.read()
major_company_splitted=major_data_raw.split("\n")
major_df=pd.DataFrame({"Companyname":major_company_splitted})
major_df.drop_duplicates(inplace=True)
major_df["level"]="major"
```

```python
with open("Specialize_Datascienc_firms.txt","r") as file:
    specialized_firm_raw=file.read()
specialized_firm_splitted=specialized_firm_raw.split("\n")
df_specialized=pd.DataFrame({"Companyname":specialized_firm_splitted})
df_specialized.drop_duplicates(inplace=True)
df_specialized["level"]="special"
```

```python
with open("medium_size.txt","r") as file:
    medium_raw=file.read()
meduim_splitted=medium_raw.split("\n")
df_medium=pd.DataFrame({"Companyname":meduim_splitted})
df_medium.drop_duplicates(inplace=True)
df_medium["level"]="medium"
```

```python
with open("Statup.txt","r") as file:
    startup_raw=file.read()
startup_splitted=startup_raw.split("\n")
df_startup=pd.DataFrame({"Companyname":startup_splitted})
df_startup.drop_duplicates(inplace=True)
df_startup["level"]="startup"
```

```python
df_company=pd.concat([df_giants,major_df,df_specialized,df_medium,df_startup])
print(len(df_company))
print(df_company["Companyname"].duplicated().sum())
df_company.drop_duplicates(subset="Companyname",inplace=True)
print(df_company["Companyname"].duplicated().sum())
```

```python
dictionary_job_level={"Entry_level":["Junior Data Engineer","Data Analyst","Buisness Intelligence(BI)",
                                     "Data Quality Analyst","Junior Data Scientist"],
                     "Mid_level":["Data Scientist","Data Engineer","Database Administrator(DBA)",
                                  "Data Architect","Buisness Intelligence (BI) Developer",
                                  "Machine Learning Engineer","Data Vizualization Specialist"],
                     "Senior_level":["Senior Data scientist","Senior Data Engineer","Lead Data Architect",
                                    "Data Consultant","Data Operation Specialist","Quantitative Analyst(quant)"],
                     "Executive_level":["Cheif Data Officer (CDO)","VP of Data","VP of Analytics","VP of Data Analysis",
                                        "Head of Data Science"]}
print(dictionary_job_level)
```
